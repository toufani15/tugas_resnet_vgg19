```python
import keras

from keras.datasets import cifar10
from keras.layers import Activation, Flatten, Conv2D, MaxPooling2D, Dense
from keras.layers import AveragePooling2D, Input, Dropout
from keras.models import Sequential, Model
from keras import utils

# import gambar
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# cek ukuran shapenya / dimensi tensornya (gambar)
# print(x_train.shape)
# print(y_train.shape)
# print(x_test.shape)
# print(y_test.shape)


# mengubah ukuran ke float32 agar mudah diproses 
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')

# normalisasi data
x_train /= 255
x_test /= 255

# membuat kelas output 10 kategori sebagai layar output/dense
num_class = 10
y_train = utils.to_categorical(y_train, num_class)
y_test = utils.to_categorical(y_test, num_class)
```

### [1] ARSITEKTUR RESIDUAL NETWORK

```python
from keras.layers import BatchNormalization

# inisialisasi input
masukan = Input(shape=(x_train.shape[1:]))

# full pre-activation block
def pre_activation_block(inputan, filterin):
    bn_1     = BatchNormalization()(inputan)
    relu_1   = Activation('relu')(bn_1)
    weight_1 = Conv2D(filterin, (3,3), padding = 'same')(relu_1)
    bn_2     = BatchNormalization()(weight_1)
    relu_2   = Activation('relu')(bn_2)
    weight_2 = Conv2D(filterin, (3,3), padding = 'same')(relu_2)
    keluaran = weight_2
    addition = keras.layers.add([inputan, keluaran]) # shape[3] (inputan harus sama dengan keluaran)
    return addition

# residual network
# kelomok 64
x = Conv2D(64, (7,7), input_shape = x_train.shape[1:], strides = (2,2), padding = 'same')(masukan)
pool = MaxPooling2D(pool_size = (2,2))(x)
blok_1_1 = pre_activation_block(inputan = pool,filterin = pool.shape.as_list()[3])
blok_1_2 = pre_activation_block(inputan = blok_1_1, filterin = blok_1_1.shape.as_list()[3])
blok_1_3 = pre_activation_block(inputan = blok_1_2, filterin = blok_1_2.shape.as_list()[3])

# kelomok 128
x1 = Conv2D(128, (3,3), strides = (2,2))(blok_1_3)
blok_2_1 = pre_activation_block(inputan = x1, filterin = x1.shape.as_list()[3])
blok_2_2 = pre_activation_block(inputan = blok_2_1, filterin = blok_2_1.shape.as_list()[3])
blok_2_3 = pre_activation_block(inputan = blok_2_2, filterin = blok_2_2.shape.as_list()[3])
blok_2_4 = pre_activation_block(inputan = blok_2_3, filterin = blok_2_3.shape.as_list()[3])

# kelomok 256
# !!! TIDAK BISA LANJUT, KARENA DIMESI OUTPUTNYA = [3,3,128],
# sehingga tidak bisa dipotong/pool lagi dimensinya karena ukurannya -
# nggak cukup
# x2 = Conv2D(256, (3,3), strides = (2,2))(blok_2_4)
# blok_3_1 = pre_activation_block(inputan = x2, filterin = x2.shape.as_list()[3])
# blok_3_2 = pre_activation_block(inputan = blok_3_1, filterin = blok_3_1.shape.as_list()[3])
# blok_3_3 = pre_activation_block(inputan = blok_3_2, filterin = blok_3_2.shape.as_list()[3])
# blok_3_4 = pre_activation_block(inputan = blok_3_3, filterin = blok_3_3.shape.as_list()[3])
# blok_3_5 = pre_activation_block(inputan = blok_3_4, filterin = blok_3_4.shape.as_list()[3])
# blok_3_6 = pre_activation_block(inputan = blok_3_5, filterin = blok_3_5.shape.as_list()[3])

# !!! TIDAK BISA LANJUT, KARENA DIMESI OUTPUTNYA = [1,1,256]
# x3 = Conv2D(512, (3,3), strides = (2,2))(blok_3_6)
# blok_4_1 = pre_activation_block(inputan = x3, filterin = x3.shape.as_list()[3])
# blok_4_2 = pre_activation_block(inputan = blok_4_1, filterin = blok_4_1.shape.as_list()[3])
# blok_4_3 = pre_activation_block(inputan = blok_4_2, filterin = blok_4_2.shape.as_list()[3])

y = AveragePooling2D(pool_size = (2,2))(blok_2_4)
luruskan_niat = Flatten()(y)
output_dense = Dense(num_class)(luruskan_niat)
aktivasi = Activation('softmax')(output_dense)
keluaran = aktivasi
model_resnet = Model(masukan, keluaran)
model_resnet.summary()

```

### [2] ARSITEKTUR VGG19

```python
input_vgg19 = Input(shape=(x_train.shape[1:]))

kon1 = Conv2D(64, (3,3), padding = 'same', activation = 'relu')(input_vgg19)
kon2 = Conv2D(64, (3,3), padding = 'same', activation = 'relu')(kon1)
pool_vgg_1 = MaxPooling2D(pool_size = (2,2))(kon2)

kon3 = Conv2D(128, (3,3), padding = 'same', activation = 'relu')(pool_vgg_1)
kon4 = Conv2D(128, (3,3), padding = 'same', activation = 'relu')(kon3)
pool_vgg_2 = MaxPooling2D(pool_size = (2,2))(kon4)

kon5 = Conv2D(256, (3,3), padding = 'same', activation = 'relu')(pool_vgg_2)
kon6 = Conv2D(256, (3,3), padding = 'same', activation = 'relu')(kon5)
kon7 = Conv2D(256, (3,3), padding = 'same', activation = 'relu')(kon6)
kon8 = Conv2D(256, (3,3), padding = 'same', activation = 'relu')(kon7)
pool_vgg_3 = MaxPooling2D(pool_size = (2,2))(kon8)

kon9 = Conv2D(512, (3,3), padding = 'same', activation = 'relu')(pool_vgg_3)
kon10 = Conv2D(512, (3,3), padding = 'same', activation = 'relu')(kon9)
kon11 = Conv2D(512, (3,3), padding = 'same', activation = 'relu')(kon10)
kon12 = Conv2D(512, (3,3), padding = 'same', activation = 'relu')(kon11)
pool_vgg_4 = MaxPooling2D(pool_size = (2,2))(kon12)



luruskan_niat_vgg = Flatten()(pool_vgg_3)
Dense_vgg = Dense(num_class)(luruskan_niat_vgg)
aktifasi_vgg = Activation('softmax')(Dense_vgg)
output_vgg19 = aktifasi_vgg
model_vgg19 = Model(input_vgg19, output_vgg19)
model_vgg19.summary()
```

### [3] ARSITEKTUR INCEPTION MODULE DIMENSIONAL REDUCTION GOOGLENET 

```python
# konstruksi CNN , arsitektur inception modul milik google

# inputs
inputs = Input(shape=(x_train.shape[1:]))
tower_1 = Conv2D(64, (1,1), padding = 'same', activation = 'relu')(inputs)
tower_1 = Conv2D(64, (3,3), padding = 'same', activation = 'relu')(tower_1)

tower_2 = Conv2D(64, (1,1), padding = 'same', activation = 'relu')(inputs)
tower_2 = Conv2D(64, (5,5), padding = 'same', activation = 'relu')(tower_2)

tower_3 = MaxPooling2D((3,3), padding = 'same', strides = (1,1))(inputs)
tower_3 = Conv2D(64, (1,1), padding = 'same', activation = 'relu')(tower_3)

x = keras.layers.concatenate([tower_1, tower_2, tower_3])

# REST net 

x = Conv2D(8, (3,3))(x)
x = Activation('relu')(x)
x = MaxPooling2D(pool_size = (2,2))(x)
x = Flatten()(x)
x = Dense(num_class)(x)
x = Activation('softmax')(x)
output = x
model = Model([inputs], output)
model.summary()
```
